```python
import pandas as pd
```

```python
df = pd.read_csv('./MODEL_IPD_coli.csv',sep=";")
```

```python
df
```

```python
ids = df["Fasta_ID"].unique()
strands = df["Strand"].unique()
```

```python
proto_json = {}

for fastaid in ids:
    for strand in strands:
        try:
            proto_json[fastaid][strand] = list(df[(df["Fasta_ID"] == fastaid) & (df["Strand"] == strand)]["Prediction"])
        except:
            proto_json[fastaid] = {}
            proto_json[fastaid][strand] = list(df[(df["Fasta_ID"] == fastaid) & (df["Strand"] == strand)]["Prediction"])
```

```python
test = pd.DataFrame(proto_json)
```

```python
test.T
```

```python
test.T.to_json('./compacted.hdf',compression="gzip")
```
